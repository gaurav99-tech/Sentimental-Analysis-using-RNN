```python
data="Food was excellent, staff was friendly and helpful. Service, Front office and housekeeping were excellent. Bar staff and security at entrance were good while receiving the guest found very professional. Hotel ambience and maintenance was excellent."
sen=sent_tokenize(data)
print(sen)
words=word_tokenize(data)
print(words)
# stop words
stopWords = set(stopwords.words('english'))
wordsFiltered = []
for w in words:
    if w not in stopWords:
        wordsFiltered.append(w)

print(wordsFiltered)
```

```python
import pandas as pd

data=pd.read_table("/content/drive/MyDrive/Data/AFINN.txt")
#print(data)
wc=[]
for x in wordsFiltered:
    for index,row in data.iterrows():
        #print(row[0],row[1])
        if x == row[0]:
            wc.append(row[1])
print(sum(wc))
```

```python
if sum(wc) > 0:
    print("Positive")
elif sum(wc) < 0:
    print("Negative")
else:
    print("Neutral")
```
